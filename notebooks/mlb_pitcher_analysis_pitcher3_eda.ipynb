### Pitch Selection

- pitch selection totals vs all batters (before vs after all star break)
- percentage change of pitch usage (before vs after all star break)
- pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters (before vs after all star break)
- pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

```python
#pitch selection totals vs all batters (before vs after all star break)

sns.countplot(x = 'pitch_type', hue = 'all_star', data = pitcher3)
plt.title('Pitch Selection Totals vs All Batters, Before and After All Star Break')
plt.ylabel('Total Pitches Thrown')
plt.xlabel('Pitch Type')
```

- Pitcher 3 increased his Four-Seam Fastball usage by 0.4% after the all star break.
- Pitcher 3 increased his Slider usage by 8.1% after the all star break.
- Pitcher 3 decreased his Curveball usage by 4.3% after the all star break.
- Pitcher 3 increased his Changeup usage by 1.2% after the all star break.
- Pitcher 3 decreased his Two-Seam Fastball usage by 1.7% after the all star break.
- We are missing data for 30 pitches of Pitcher 3's pitch data, but chose not to remove it in our data cleaning phase due to the importance that other columns in those rows may have on our analysis.

```python
#pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters 
# (before vs after all star break)

pitcher3.groupby(['bats', 'all_star']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL', 'pitch_type_FT']].plot(kind = 'barh', figsize = (16,10), 
                                               title = 'Pitch Selection Percentage vs Left/Right Hand Batters, Before and After All Star Break')
```

- Pitcher 3 decreased his Curveball usage against left-handed batters after the all star break. Pitcher 2 completely removed his Curveball from his arsenal against right-handed batters after the all star break.
- Pitcher 3 decreased his Four-Seam Fastball usage against both right-handed and left-handed batters after the all star break.
- Pitcher 3 decreased his Changeup usage against right-handed batters after the all star break. However, he increased his Changeup usage against left-handed batters after the all star break.
- Pitcher 3 increased his Slider usage against both right-handed and left-handed batters after the all star break.
- Pitcher 3 decreased his Two-Seam Fastball usage against both right-handed batters and left-handed batters after the all star break.

## Key Findings, Conclusions, Recommendations

- Pitcher 3 also seemed to have mostly optimized his pitch selection. Pitcher 3 had his highest success rate with his Four-Seam Fastball among the pitches he threw a lot of before the all-star break, while his Changeup and Slider were less inspiring. He does not use his Curveball and Two-Seam Fastball enough for us to worry about, although he seems to do well with both of them in a pinch, especially the Curveball. Pitcher 3 increased the usage of his favored breaking ball, the Slider, with over an 8% increase in usage. This helped to offset the over-usage of his fastball, which he throws at over a 50% clip. The higher usage of his Slider also made his Four-Seam Fastball more effective after the all-star break, with almost 7% increased called strike percentage, while also improving his called strike percentage of his Slider (it could still use more improvement though). He could benefit from improving a third pitch because his Changeup, which he increased his usage of, is not very effective.
- Perhaps the third pitch that Pitcher 3 should use more of is his Curveball, which has a high spin rate and seemingly good pitch movement. As referenced above, it is a pitch hes thrown pretty well when hes actually thrown it, which he doesnt do very often.
- Like Pitcher 2, one item that stood out among the physical attributes of Pitcher 3s pitching was the large decrease in his horizontal release point (release_x) for all of his pitches, which suggests that perhaps he is threw from a better arm slot after the all-star break.
- Pitcher 3 had his highest called strike percentage when Catcher 10 caught for him. There is not a major drop-off to Catcher 4 though. With umps, there is not much fluctuation with called strike percentage than Pitchers 1 and 2, suggesting he might have better luck with umps. Pitcher 3 had his best called strike percentage with Ump 50, and his lowest with Ump 45.
