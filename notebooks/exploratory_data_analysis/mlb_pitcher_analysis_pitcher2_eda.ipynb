### Pitch Selection

- pitch selection totals vs all batters (before vs after all star break)
- percentage change of pitch usage (before vs after all star break)
- pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters (before vs after all star break)
- pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

```python
#pitch selection totals vs all batters (before vs after all star break)

sns.countplot(x = 'pitch_type', hue = 'all_star', data = pitcher2)
plt.title('Pitch Selection Totals vs All Batters, Before and After All Star Break')
plt.ylabel('Total Pitches Thrown')
plt.xlabel('Pitch Type')
```

- Pitcher 2 increased his Four-Seam Fastball usage by 1.1% after the all star break.
- Pitcher 2 increased his Slider usage by 2.1% after the all star break.
- Pitcher 2 decreased his Curveball usage by 3.3% after the all star break.
- Pitcher 2 increased his Changeup usage by 1.7% after the all star break.
- Pitcher 2 decreased his Two-Seam Fastball usage by 1.7% after the all star break.

```python
#pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters 
# (before vs after all star break)

pitcher2.groupby(['bats', 'all_star']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL', 'pitch_type_FT']].plot(kind = 'barh', figsize = (16,10), 
                                               title = 'Pitch Selection Percentage vs Left/Right Hand Batters, Before and After All Star Break')
```

- Pitcher 2 increased his Curveball usage against both right-handed and left-handed batters after the all star break.
- Pitcher 2 increased his Changeup usage against both right-handed and left-handed batters after the all star break.
- Pitcher 2 increased his Four-Seam Fastball usage against right-handed batters after the all star break. However, he decreased his Four-Seam Fastball usage against left-handed batters after the all star break.
- Pitcher 2 increased his Slider usage against both right-handed and left-handed batters after the all star break.
- Pitcher 2 decreased his Two-Seam Fastball usage against both right-handed batters and left-handed batters after the all star break.

### Pitch Success Rate
- pitch success rate (called strike percentage)

```python
#pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

pitcher2.groupby(['all_star', 'pitch_result']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL']].plot(kind = 'barh', figsize = (15,10), 
                              title = 'Pitch Selection Percentage vs All Batters, Strikes vs Balls, Before and After All Star Break')
```

- Pitcher 2 had an increased success rate (called strike percentage) with his Four-Seam Fastball, Slider, and Curveball after the all star break.
- Pitcher 2 had a decreased success rate (called strike percentage) with his Changeup and Two-Seam Fastball after the all star break. However, this is not significant for his Two-Seam Fastball because he hardly threw that pitch in the second half of the season.

### Pitch Location
- pitch location by pitch type (before vs after all star break)

- Pitcher 2 does not have good command of his Changeup and has trouble locating the strike zone with it. He uses this pitch enough to warrant concern.
- When Pitcher 2 does miss the strike zone with his Slider, he tends to miss the bottom right corner of the zone the most in comparison to any other parts of the zone.
- When Pitcher 2 does miss the strike zone with his Curveball, he tends to miss the bottom right corner of the zone and the top left corner of the zone more than any other parts of the zone.

## Key Findings, Conclusions, Recommendations

- Pitcher 2 seemed to have optimized his pitch selection after the all-star break. Pitcher 2 had his highest success rate with his Four-Seam Fastball and breaking ball combination of his Slider and Curveball before the all-star break. He increased his usage of his Four-Seam Fastball and Slider, which is justified by the fact that his success rate with both after the all-star break went up, especially with the Four-Seam Fastball where he saw a near 6% increase in called strike percentage. Pitcher 2 also saw an increase in his Curveball success rate, despite using less of it after the all-star break. Pitcher 2 also upped his usage of his Changeup after the all-star break, which is not recommended given the drop in his success rate in the pitch after the all-star break. Pitcher 2 does not use his Two-Seam Fastball enough for us to believe it has much significance.
- One item that stood out among the physical attributes of Pitcher 2s pitching was the medium increase in his horizontal release point (release_x) for all of his pitches, which suggests that perhaps he is threw from a better arm slot after the all-star break.
- Another item of note was that Pitcher 2 had a large increase in his Changeup spin rate, which is interesting considering he did not throw it very successfully after the all-star break. Perhaps this suggests that we could maximize this pitch more by ensuring he uses this pitch in the proper situations, and also in the proper locations in the strike zone.
- In terms of pitch location, Pitcher 2 does not have good command of his Changeup and has trouble locating the strike zone with it. He uses this pitch enough to warrant concern.
- Pitcher 2 had his highest called strike percentage when Catcher 2 and Catcher 7 caught for him. There is not a major drop-off to Catcher 3 either. Pitcher 2 had his lowest called strike percentage when Catcher 9 caught for him. With umps, there is slightly less fluctuation with called strike percentage than Pitcher 1. Pitcher 2 had his best called strike percentage with Ump 26, and his lowest with Ump 49.
