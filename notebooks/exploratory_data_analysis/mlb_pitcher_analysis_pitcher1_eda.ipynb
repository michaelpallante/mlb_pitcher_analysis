### Pitch Selection
- pitch selection totals vs all batters (before vs after all star break)
- percentage change of pitch usage (before vs after all star break)
- pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters (before vs after all star break)
- pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

```python
#pitch selection totals vs all batters (before vs after all star break)

sns.countplot(x = 'pitch_type', hue = 'all_star', data = pitcher1)
plt.title('Pitch Selection Totals vs All Batters, Before and After All Star Break')
plt.ylabel('Total Pitches Thrown')
plt.xlabel('Pitch Type')
```

- Pitcher 1 decreased his Four-Seam Fastball usage by 1.4% after the all star break.
- Pitcher 1 decreased his Slider usage by 4.1% after the all star break.
- Pitcher 1 increased his Curveball usage by 2.4% after the all star break.
- Pitcher 1 increased his Changeup usage by 2.2% after the all star break.
- Pitcher 1 increased his Cutter usage by 0.9% after the all star break.
- Pitcher 1 completely got rid of his Two-Seam Fastball from his arsenal after the all star break. It was a pitch he only used once before the all star break anyway, so this is insignificant.

```python
#pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters 
# (before vs after all star break)

pitcher1.groupby(['bats', 'all_star']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL', 'pitch_type_FT']].plot(kind = 'barh', figsize = (16,10), 
                                               title = 'Pitch Selection Percentage vs Left/Right Hand Batters, Before and After All Star Break')
```

- Pitcher 1 increased his Curveball usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 increased his Changeup usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 increased his Cutter usage against right-handed batters after the all star break. However, he decreased his Cutter usage against left-handed batters after the all star break.
- Pitcher 1 decreased his Four-Seam Fastball usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 increased his Slider usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 decreased his Two-Seam Fastball usage against right-handed batters after the all star break. He did not use this pitch against left-handed batters at all.

### Pitch Success Rate
- pitch success rate (called strike percentage)

```python
#pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

pitcher1.groupby(['all_star', 'pitch_result']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL']].plot(kind = 'barh', figsize = (15,10), 
                              title = 'Pitch Selection Percentage vs All Batters, Strikes vs Balls, Before and After All Star Break')
```

- Pitcher 1 had a decreased success rate (called strike percentage) with all of his pitches except for his Changeup after the all star break. This suggests that perhaps he should revisit the amount he uses each pitch. We also have to examine all other variables to see if they also had an impact on his success rate in the second half of the season. 

### Pitch Location
- pitch location by pitch type (before vs after all star break)

- Pitcher 1 does not have good command of his Changeup or Cutter and has trouble locating the strike zone with both. Fortunately, he does not use his Cutter often. However, this is more of an issue with his Changeup because he does use it enough to warrant concern.
- When Pitcher 1 does miss the strike zone with his Slider, he tends to miss the bottom right corner of the zone the most in comparison to any other parts of the zone.
- When Pitcher 1 does miss the strike zone with his Curveball, he tends to miss the bottom right corner of the zone and to the left of the zone more than any other parts of the zone.

## Key Findings, Conclusions, Recommendations

- Pitcher 1 had his highest success rate (called strike percentage) with his Four-Seam Fastball/Slider/Curveball combination before the all-star break. His decrease in percent change in usage of his Slider and even his Four-Seam Fastball are not recommended. He should place more emphasis on throwing these pitches as opposed to his Changeup and Cutter, which have both been rather ineffective pitches for him the entire season, according to his low success rate with both pitches. He was wise in increasing in Curveball usage though, as this was his most effective pitch before the all-star break.
- However, it is important to note that even his most effective pitches dropped in effectiveness after the all-star break, which suggests that there is more to his story than the above. There were no physical attributes of his pitching that were of major concern, although we did see some fairly significant fluctuation with his horizontal pitch movement (pfx_x). His Curveball and Slider movement actually seemed to improve, while his other pitches did not.
- In terms of pitch location, Pitcher 1 does not have good command of his Changeup or Cutter and has trouble locating the strike zone with both. Fortunately, he does not use his Cutter often. However, this is more of an issue with his Changeup because he does use it enough to warrant concern.
- There is not a drastic difference in terms of which catcher is catching for Pitcher 1. He has a slightly higher called strike percentage with Catcher 6, than with Catcher 1. There is some pretty significant fluctuation with his called strike percentage due to umps though. He earned over 30+% better called strike percentage with some umps than he did with others, perhaps suggesting he could have some bad luck there. Pitcher 1 had his best called strike percentage with Ump 13, and his lowest with Ump 8.
- Overall, it seems like pitch selection and some bad luck factors are the reason for Pitcher 1’s ERA inflating after the all-star break. It seems like there is more to this story that we could decipher with additional data.
