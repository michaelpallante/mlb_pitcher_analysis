#### Pitch Selection

- pitch selection totals vs all batters (before vs after all star break)
- percentage change of pitch usage (before vs after all star break)
- pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters (before vs after all star break)
- pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

```python
#pitch selection totals vs all batters (before vs after all star break)

sns.countplot(x = 'pitch_type', hue = 'all_star', data = pitcher2)
plt.title('Pitch Selection Totals vs All Batters, Before and After All Star Break')
plt.ylabel('Total Pitches Thrown')
plt.xlabel('Pitch Type')
```

- Pitcher 2 increased his Four-Seam Fastball usage by 1.1% after the all star break.
- Pitcher 2 increased his Slider usage by 2.1% after the all star break.
- Pitcher 2 decreased his Curveball usage by 3.3% after the all star break.
- Pitcher 2 increased his Changeup usage by 1.7% after the all star break.
- Pitcher 2 decreased his Two-Seam Fastball usage by 1.7% after the all star break.

```python
#pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters 
# (before vs after all star break)

pitcher2.groupby(['bats', 'all_star']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL', 'pitch_type_FT']].plot(kind = 'barh', figsize = (16,10), 
                                               title = 'Pitch Selection Percentage vs Left/Right Hand Batters, Before and After All Star Break')
```

- Pitcher 2 increased his Curveball usage against both right-handed and left-handed batters after the all star break.
- Pitcher 2 increased his Changeup usage against both right-handed and left-handed batters after the all star break.
- Pitcher 2 increased his Four-Seam Fastball usage against right-handed batters after the all star break. However, he decreased his Four-Seam Fastball usage against left-handed batters after the all star break.
- Pitcher 2 increased his Slider usage against both right-handed and left-handed batters after the all star break.
- Pitcher 2 decreased his Two-Seam Fastball usage against both right-handed batters and left-handed batters after the all star break.

#### Pitch Success Rate
- pitch success rate (called strike percentage)

```python
#pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

pitcher2.groupby(['all_star', 'pitch_result']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL']].plot(kind = 'barh', figsize = (15,10), 
                              title = 'Pitch Selection Percentage vs All Batters, Strikes vs Balls, Before and After All Star Break')
```

- Pitcher 2 had an increased success rate (called strike percentage) with his Four-Seam Fastball, Slider, and Curveball after the all star break.
- Pitcher 2 had a decreased success rate (called strike percentage) with his Changeup and Two-Seam Fastball after the all star break. However, this is not significant for his Two-Seam Fastball because he hardly threw that pitch in the second half of the season.

#### Physical Attributes of Pitch

- spin rate by pitch type (before vs after all star break)
- release velocity by pitch type (before vs after all star break)
- horizontal release point (release_x) by pitch type (before vs after all star break)
- release point distance from home plate (release_y) by pitch type (before vs after all star break)
- vertical release point (release_z) by pitch type (before vs after all star break)
- horizontal pitch movement (pfx_x) by pitch type (before vs after all star break)
- vertical pitch movement (pfx_z) by pitch type (before vs after all star break)
- extension by pitch type (before vs after all star break)

```python
#spin rate by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'spin_rate', hue = 'all_star', data = pitcher2)
plt.title('Spin Rate by Pitch Type, Before and After All Star Break')
plt.ylabel('Spin Rate')
plt.xlabel('Pitch Type')
```

- Pitcher 2 had a large increase in his Changeup spin rate after the all star break.
- Pitcher 2 had a very slight increase in his Slider spin rate after the all star break.
- Pitcher 2 had a very slight decrease in his Curveball spin rate after the all star break.
- Pitcher 2 had a very slight increase in his Four-Seam Fastball spin rate after the all star break.
- Pitcher 2 had a medium increase in his Two-Seam Fastball spin rate after the all star break.

```python
#release velocity by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_velo', hue = 'all_star', data = pitcher2)
plt.title('Release Velocity by Pitch Type, Before and After All Star Break')
plt.ylabel('Release Velocity')
plt.xlabel('Pitch Type')
```

- Pitcher 2 did not see a significant change in his release velocity for his Changeup and Slider after the all star break.
- Pitcher 2 had a very slight increase in his release velocity for his Curveball, Four-Seam Fastball, and Two-Seam Fastball after the all star break.

```python
#horizontal release point (release_x) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_x', hue = 'all_star', data = pitcher2)
plt.title('Release_X by Pitch Type, Before and After All Star Break')
plt.ylabel('Release_X')
plt.xlabel('Pitch Type')
```

- Pitcher 2 had a medium increase in his horizontal release point for all of his pitches after the all star break.

```python
#release point distance from home plate (release_y) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_y', hue = 'all_star', data = pitcher2)
plt.title('Release_Y by Pitch Type, Before and After All Star Break')
plt.ylabel('Release_Y')
plt.xlabel('Pitch Type')
```

- Pitcher 2 did not see a significant change in his release point distance from home plate for any of his pitches after the all star break.

```python
#vertical release point (release_z) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_z', hue = 'all_star', data = pitcher2)
plt.title('Release_Z by Pitch Type, Before and After All Star Break')
plt.ylabel('Release_Z')
plt.xlabel('Pitch Type')
```

- Pitcher 2 had a very slight decrease in his vertical release point for all of his pitches after the all star break.

```python
#horizontal pitch movement (pfx_x) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'pfx_x', hue = 'all_star', data = pitcher2)
plt.title('PFX_X by Pitch Type, Before and After All Star Break')
plt.ylabel('PFX_X')
plt.xlabel('Pitch Type')
```

- Pitcher 2 had a large decrease in his horizontal pitch movement for his Curveball after the all star break.
- Pitcher 2 had a slight decrease in his horizontal pitch movement for his Four-Seam Fastball and his Changeup after the all star break.
- Pitcher 2 had a slight increase in his horizontal pitch movement for his Two-Seam Fastball and his Slider after the all star break.

```python
#vertical pitch movement (pfx_z) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'pfx_z', hue = 'all_star', data = pitcher2)
plt.title('PFX_Z by Pitch Type, Before and After All Star Break')
plt.ylabel('PFX_Z')
plt.xlabel('Pitch Type')
```

- Pitcher 2 had a medium decrease in his vertical pitch movement for his Slider after the all star break.
- Pitcher 2 had a medium increase in his vertical pitch movement for his Curveball and Two-Seam Fastball after the all star break.
- Pitcher 2 did not see a significant change in his vertical pitch movement for his Four-Seam Fastball and his Changeup after the all star break.

```python
#extension by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'extension', hue = 'all_star', data = pitcher2)
plt.title('Extension by Pitch Type, Before and After All Star Break')
plt.ylabel('Extension')
plt.xlabel('Pitch Type')
```

- Pitcher 2 did not have any significant changes in his extension for all of his pitches after the all star break.
