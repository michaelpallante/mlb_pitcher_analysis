#### Pitch Selection
- pitch selection totals vs all batters (before vs after all star break)
- percentage change of pitch usage (before vs after all star break)
- pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters (before vs after all star break)
- pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

```python
#pitch selection totals vs all batters (before vs after all star break)

sns.countplot(x = 'pitch_type', hue = 'all_star', data = pitcher1)
plt.title('Pitch Selection Totals vs All Batters, Before and After All Star Break')
plt.ylabel('Total Pitches Thrown')
plt.xlabel('Pitch Type')
```

- Pitcher 1 decreased his Four-Seam Fastball usage by 1.4% after the all star break.
- Pitcher 1 decreased his Slider usage by 4.1% after the all star break.
- Pitcher 1 increase his Curveball usage by 2.4% after the all star break.
- Pitcher 1 increased his Changeup usage by 2.2% after the all star break.
- Pitcher 1 increased his Cutter usage by 0.9% after the all star break.
- Pitcher 1 completely got rid of his Two-Seam Fastball from his arsenal after the all star break. It was a pitch he only used once before the all star break anyway, so this is insignificant.

```python
#pitch selection percentage vs left-handed batters, pitch selection vs right-handed batters 
# (before vs after all star break)

pitcher1.groupby(['bats', 'all_star']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL', 'pitch_type_FT']].plot(kind = 'barh', figsize = (16,10), 
                                               title = 'Pitch Selection Percentage vs Left/Right Hand Batters, Before and After All Star Break')
```

- Pitcher 1 increased his Curveball usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 increased his Changeup usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 increased his Cutter usage against right-handed batters after the all star break. However, he decreased his Cutter usage against left-handed batters after the all star break.
- Pitcher 1 decreased his Four-Seam Fastball usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 increased his Slider usage against both right-handed and left-handed batters after the all star break.
- Pitcher 1 decreased his Two-Seam Fastball usage against right-handed batters after the all star break. He did not use this pitch against left-handed batters at all.

#### Pitch Success Rate
- pitch success rate (called strike percentage)

```python
#pitch selection percentage vs all batters, strikes vs balls (before vs after all star break)

pitcher1.groupby(['all_star', 'pitch_result']).mean()[['pitch_type_CB',
       'pitch_type_CH', 'pitch_type_CT', 'pitch_type_FF',
       'pitch_type_SL']].plot(kind = 'barh', figsize = (15,10), 
                              title = 'Pitch Selection Percentage vs All Batters, Strikes vs Balls, Before and After All Star Break')
```

- Pitcher 1 had a decreased success rate (called strike percentage) with all of his pitches except for his Changeup after the all star break. This suggests that perhaps he should revisit the amount he uses each pitch. We also have to examine the other variables to see if they also had an impact on his success rate in the second half of the season.

#### Physical Attributes of Pitch

- spin rate by pitch type (before vs after all star break)
- release velocity by pitch type (before vs after all star break)
- horizontal release point (release_x) by pitch type (before vs after all star break)
- release point distance from home plate (release_y) by pitch type (before vs after all star break)
- vertical release point (release_z) by pitch type (before vs after all star break)
- horizontal pitch movement (pfx_x) by pitch type (before vs after all star break)
- vertical pitch movement (pfx_z) by pitch type (before vs after all star break)
- extension by pitch type (before vs after all star break)

```python
#spin rate by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'spin_rate', hue = 'all_star', data = pitcher1)
plt.title('Spin Rate by Pitch Type, Before and After All Star Break')
plt.ylabel('Spin Rate')
plt.xlabel('Pitch Type')
```

- Pitcher 1 had a slight decrease in his Changeup spin rate after the all star break.
- Pitcher 1 had a slight increase in his Slider spin rate after the all star break.
- Pitcher 1 had a marginal increase in his Curveball spin rate after the all star break.
- Pitcher 1 had a very slight decrease in his Four-Seam Fastball spin rate after the all star break.
- Pitcher 1 had a very slight decrease in his Cutter spin rate after the all star break.
- Pitcher 1 did not have a change in his Two-Seam Fastball spin rate after the all star break because he abandoned that pitch after only using it once before the all star break.

```python
#release velocity by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_velo', hue = 'all_star', data = pitcher1)
plt.title('Release Velocity by Pitch Type, Before and After All Star Break')
plt.ylabel('Release Velocity')
plt.xlabel('Pitch Type')
```

- Pitcher 1 did not see a significant change in his release velocity for any of his pitches after the all star break, except for a very slight increase in his Curveball release velocity.
- Pitcher 1 did not have a change in his Two-Seam Fastball release velocity after the all star break because he abandoned that pitch after only using it once before the all star break.

```python
#horizontal release point (release_x) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_x', hue = 'all_star', data = pitcher1)
plt.title('Release_X by Pitch Type, Before and After All Star Break')
plt.ylabel('Release_X')
plt.xlabel('Pitch Type')
```

```python
#release point distance from home plate (release_y) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_y', hue = 'all_star', data = pitcher1)
plt.title('Release_Y by Pitch Type, Before and After All Star Break')
plt.ylabel('Release_Y')
plt.xlabel('Pitch Type')
```

- Pitcher 1 did not have a change in his Two-Seam Fastball release point distance from home plate after the all star break because he abandoned that pitch after only using it once before the all star break.

```python
#vertical release point (release_z) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'release_z', hue = 'all_star', data = pitcher1)
plt.title('Release_Z by Pitch Type, Before and After All Star Break')
plt.ylabel('Release_Z')
plt.xlabel('Pitch Type')
```

```python
#horizontal pitch movement (pfx_x) by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'pfx_x', hue = 'all_star', data = pitcher1)
plt.title('PFX_X by Pitch Type, Before and After All Star Break')
plt.ylabel('PFX_X')
plt.xlabel('Pitch Type')
```

```python
#extension by pitch type (before vs after all star break)

sns.barplot(x = 'pitch_type', y = 'extension', hue = 'all_star', data = pitcher1)
plt.title('Extension by Pitch Type, Before and After All Star Break')
plt.ylabel('Extension')
plt.xlabel('Pitch Type')
```
